### 비슷한 칼럼이 동일한지 확인

```python
# 두 칼럼이 동일한지 확인
is_equal = df['Fwd Header Length'] == df['Fwd Header Length.1']

# 동일한 값이 있는지 여부 출력
if is_equal.all():
    print("두 칼럼 'Fwd Header Length'와 'Fwd Header Length.1'은 동일합니다.")
else:
    print("두 칼럼 'Fwd Header Length'와 'Fwd Header Length.1'은 동일하지 않습니다.")
    
    # 동일하지 않은 인덱스 출력
    print("동일하지 않은 인덱스:", is_equal[is_equal == False].index)

# 0.0s
```
